This backend lets you communicate directly with MonarQ.

## Default usage

Here is a typical workflow for using the ```MonarQBackend```: 

0. First and foremost, you have to make sure to have python version ```3.10.x``` intalled

1. install the plugin (refer to the README)

2. Import dependencies:

````python
# this line imports the pennylane library. It contains everything necessary for quantum programming
import qiskit

# this line imports the client class which is required for authenticating with MonarQ
from qiskit_calculquebec.API.client import MonarqClient

# This line imports the MonarQBackend class
from qiskit_calculquebec.backends import MonarQBackend
````

3. create a client for your device

````python
# This is how you create a client. Change the values in the parentheses for your credentials
# project is optional. It will be set to default if you don't specify it
my_client = MonarqClient("your host", "your user", "your access token", "your project")
````

4. create a device using your client

There are 1 mandatory arguments for the backend : 
- a client

And one optionnal :
- The name of the machine you want to use "yukon" or "monarq" (MonarQ is the default) 

````python
# Create your backend
backend = MonarQBackend(client=my_client)
````

5. create your circuit

````python
qc = qiskit.QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
````

For more information  about qiskit operations, click [here](https://quantum.cloud.ibm.com/docs/en/api/qiskit/circuit_library)

6. Transpile and run your circuit on MonarQ

````python

compiled_circuit = qiskit.transpile(qc, backend=backend)
job = backend.run(compiled_circuit, shots=1024)
result = job.result()
print(result.get_counts())
````
